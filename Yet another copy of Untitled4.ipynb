Introduction the Iris dataset is a well-known dataset used in data science and machine learning for classification tasks. It contains measurements of iris flowers, including sepal length, sepal width, petal length, and petal width. The dataset includes three species of iris flowers: Setosa, Versicolor, and Virginica.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("iris")
df.head()
```

```python
print(df.shape)
print(df.columns)
df.head()
```

The dataset contains 150 rows and 5 columns.The columns include sepal length,sepal width,petal length,petal width,and species.


```python
df.isnull().sum()
df.duplicated().sum()
```

```python
df.info()
df.describe()
```

```python
sns.scatterplot(x="sepal_length",y="petal_length",hue="species",data=df)
plt.title("sepal length vs petal length")
plt.show()
```

The scatter plot shows the relationship between sepal length and petal length.Setosa species is clearly separated from the other species.

```python
plt.hist(df["sepal_length"])
plt.title("Distribution of sepal length")
plt.show()
```

The histogram shows the distribution of sepal length values.Most values lie between 5 and 7.


```python
sns.boxplot(x="species",y="sepal_length",data=df)
plt.title("Box plot of sepal length")
plt.show()
```

The boxplot shows the spread of sepal length for each species and helps identify outliers.
